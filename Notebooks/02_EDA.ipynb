```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Se cargan los archivos limpios
df_homicidios = pd.read_csv("../Datasets/homicidios_limpio.csv")
df_lesiones = pd.read_csv("../Datasets/lesiones_limpio.csv")
```

```python
# Coventir a formato fecha
df_homicidios['fecha'] = pd.to_datetime(df_homicidios['fecha'], errors='coerce')
df_lesiones['fecha'] = pd.to_datetime(df_lesiones['fecha'], errors='coerce')
```

```python
# Convertir a tipo numérico con downcast a integer
df_homicidios['franja_horaria'] = pd.to_numeric(df_homicidios['franja_horaria'], errors='coerce')
df_lesiones['franja_horaria'] = pd.to_numeric(df_lesiones['franja_horaria'], errors='coerce')

df_homicidios['comuna'] = pd.to_numeric(df_homicidios['comuna'], errors='coerce')
df_lesiones['comuna'] = pd.to_numeric(df_lesiones['comuna'], errors='coerce')

df_homicidios['edad'] = pd.to_numeric(df_homicidios['edad'], errors='coerce')
df_lesiones['edad'] = pd.to_numeric(df_lesiones['edad'], errors='coerce')
```

```python
# Lista de columnas a categorizar
categorical_columns = ['tipo_de_calle', 'vehiculo_victima', 'vehiculo_acusado', 'rol', 'sexo','gravedad']

# Iterar sobre las columnas y categorizar solo si están presentes en df_homicidios
for column in categorical_columns:
    if column in df_homicidios.columns:
        df_homicidios[column] = df_homicidios[column].astype('category')

# Iterar sobre las columnas y categorizar solo si están presentes en df_lesiones
for column in categorical_columns:
    if column in df_lesiones.columns:
        df_lesiones[column] = df_lesiones[column].astype('category')
```

**Homicidios:**<br>
En el conjunto de datos sobre homicidios, se detectan ausencias de datos. Existen registros incompletos en la columna de "franja_horaria", así como en las columnas que detallan los vehículos implicados ("vehiculo_victima" y "vehiculo_acusado"), las funciones de los individuos ("rol"), y detalles demográficos como el sexo y la edad ("sexo" y "edad"). Estas omisiones pueden atribuirse a varios factores, incluyendo fallos al recoger los datos, supresiones deliberadas o la inexistencia de información en determinadas instancias.

**Lesiones:**<br>
Respecto al conjunto de datos de lesiones, se hallan datos incompletos en las columnas que describen la posición geográfica ("longitud" y "latitud"), los vehículos involucrados ("vehiculo_victima" y "vehiculo_acusado"), así como la comuna, el tipo de vía y la franja horaria.

Cada columna de estos datos posee potencialmente información crucial acerca de los incidentes. Descartar las columnas con datos faltantes podría conllevar a la eliminación de información esencial para entender la naturaleza y circunstancias de los accidentes, así como otros elementos significativos. Además, esta acción podría introducir distorsiones analíticas, dado que los datos remanentes quizás no reflejen fielmente la realidad de los siniestros viales, lo cual tendría un impacto negativo en las conclusiones y recomendaciones resultantes del estudio.

```python
# Identificar registros duplicados en el dataframe de homicidios
duplicados_homicidios = df_homicidios[df_homicidios.duplicated()]
duplicados_homicidios
```

```python
# Identificar registros duplicados en el dataframe de lesiones
duplicados_lesiones = df_lesiones[df_lesiones.duplicated()]
columnas_duplicados_lesiones = duplicados_lesiones["id_siniestro"]
columnas_duplicados_lesiones
```

```python
lesiones_consulta = "LC-2019-0009807"
consulta_resultados = df_lesiones[df_lesiones['id_siniestro'] == lesiones_consulta]
consulta_resultados
```

Los registros repetidos bajo el "id_siniestro" ocurren porque cada siniestro puede involucrar a múltiples víctimas. Cada una de estas víctimas se representa en una fila separada, compartiendo el mismo "id_siniestro". Para resolver esta situación y asegurar la distinción entre los registros, se asignará un 'id_victima' único a cada fila.

```python
# Agregar una nueva columna para el identificador único de la víctima
df_lesiones['id_victima'] = df_lesiones.groupby('id_siniestro').cumcount() + 1

# Mostrar el DataFrame resultante
df_lesiones.head()
```

### Accidentes por comuna

```python
# Se crea un DataFrame que contenga la cantidad de accidentes por comuna en df_lesiones
lesiones_por_comuna = df_lesiones['comuna'].value_counts().reset_index()
lesiones_por_comuna.columns = ['comuna', 'cantidad_lesiones']

# Se crea un DataFrame que contenga la cantidad de accidentes por comuna en df_homicidios
homicidios_por_comuna = df_homicidios['comuna'].value_counts().reset_index()
homicidios_por_comuna.columns = ['comuna', 'cantidad_homicidios']

# Se unen ambos DataFrames en base a la comuna
df_comparacion_comunas = lesiones_por_comuna.merge(homicidios_por_comuna, on='comuna', how='outer')

# Se ordenan el DataFrame por comuna
df_comparacion_comunas = df_comparacion_comunas.sort_values(by='comuna')

# Se crean los subgráficos
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Barra para lesiones
axs[0].bar(df_comparacion_comunas['comuna'], df_comparacion_comunas['cantidad_lesiones'], color='blue')
axs[0].set_title('Lesiones por Comuna')
axs[0].set_xlabel('Comuna')
axs[0].set_ylabel('Cantidad de Lesiones')

# Barra para homicidios
axs[1].bar(df_comparacion_comunas['comuna'], df_comparacion_comunas['cantidad_homicidios'], color='red')
axs[1].set_title('Homicidios por Comuna')
axs[1].set_xlabel('Comuna')
axs[1].set_ylabel('Cantidad de Homicidios')

plt.annotate("""Se destaca que la Comuna 1 presenta la mayor incidencia de incidentes, seguido de las comunas 4 y 9,
             tanto en términos de lesionados como de víctimas fatales.""",
            xy = (0, -0.2), xycoords='axes fraction', ha='center', va="center", fontsize=12)
#plt.tight_layout()
plt.show()
```

```python
# Se crea los gráficos de barras uno al lado del otro
plt.figure(figsize=(15, 6))

# Gráfico de lesiones por comuna
plt.subplot(1, 2, 1)
sns.countplot(x='comuna', data=df_lesiones, palette='colorblind', hue=df_homicidios['nro_victimas'])
plt.title('Lesiones por Comuna (desagregado)')
plt.xlabel('Comuna')
plt.ylabel('Cantidad de Lesiones')

# Gráfico de homicidios por comuna
plt.subplot(1, 2, 2)
sns.countplot(x='comuna', data=df_homicidios, palette='colorblind', hue=df_homicidios['nro_victimas'])
plt.title('Homicidios por Comuna (desagregado)')
plt.xlabel('Comuna')
plt.ylabel('Cantidad de Homicidios')

plt.annotate("""Se aprecia que principalmente solo se registra una única víctima tanto en lesiones como homicidios""",
            xy = (0, -0.2), xycoords='axes fraction', ha='center', va="center", fontsize=12)
#plt.tight_layout()
plt.show()
```

### Linea de tiempo

```python
# Agrupar por año y mes, y sumar el número de víctimas
rel_tiempo_df_lesiones = df_lesiones.groupby(['anho', 'mes'])['nro_victimas'].sum().reset_index()
rel_tiempo_df_homicidios = df_homicidios.groupby(['anho', 'mes'])['nro_victimas'].sum().reset_index()
```

```python
#Para la realización del gráfico se transforman los datos a tipo int
rel_tiempo_df_lesiones["anho"] = rel_tiempo_df_lesiones["anho"].astype("Int64")
rel_tiempo_df_lesiones["mes"] = rel_tiempo_df_lesiones["mes"].astype("Int64")

rel_tiempo_df_homicidios["anho"] = rel_tiempo_df_homicidios["anho"].astype("Int64")
rel_tiempo_df_homicidios["mes"] = rel_tiempo_df_homicidios["mes"].astype("Int64")
```

```python
# Combinar AÑO y MES para obtener una sola variable de tiempo, creando la columna "Fecha" como str
rel_tiempo_df_lesiones["fecha"] = rel_tiempo_df_lesiones["anho"].astype(str) + '-' + rel_tiempo_df_lesiones["mes"].astype(str)

# Convertir la variable de tiempo a formato datetime
rel_tiempo_df_lesiones["fecha"] = pd.to_datetime(rel_tiempo_df_lesiones["fecha"])

# Ordenar el DataFrame por fecha para que tenga orden cronológico
rel_tiempo_df_lesiones = rel_tiempo_df_lesiones.sort_values("fecha")


# Combinar AÑO y MES para obtener una sola variable de tiempo, creando la columna "Fecha" como str
rel_tiempo_df_homicidios["fecha"] = rel_tiempo_df_homicidios["anho"].astype(str) + '-' + rel_tiempo_df_homicidios["mes"].astype(str)

# Convertir la variable de tiempo a formato datetime
rel_tiempo_df_homicidios["fecha"] = pd.to_datetime(rel_tiempo_df_homicidios["fecha"])

# Ordenar el DataFrame por fecha para que tenga orden cronológico
rel_tiempo_df_homicidios = rel_tiempo_df_homicidios.sort_values("fecha")
```

```python
# Se crea los gráficos de barras uno al lado del otro
plt.figure(figsize=(15, 6))

# Gráfico de linea de tiempo para lesiones
plt.subplot(1, 2, 1)
plt.plot(rel_tiempo_df_lesiones["fecha"], rel_tiempo_df_lesiones["nro_victimas"], marker='o', color='skyblue')
plt.title('Linea de tiempo - lesiones')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de lesiones')
plt.xticks(rotation=45)

# Gráfico de linea de tiempo para homicidios 
plt.subplot(1, 2, 2)
plt.plot(rel_tiempo_df_homicidios["fecha"], rel_tiempo_df_homicidios["nro_victimas"], marker='o', color='salmon')
plt.title('Linea de tiempo - homicidios')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de homicidios')
plt.xticks(rotation=45)

plt.annotate("""Se observa la drástica disminución de lesiones y homicidios en épocas de cuarentenas (2020),
             volviendo a subir luego de esta aunque sin regresar a niveles pre-pandemia""",
            xy = (0, -0.3), xycoords='axes fraction', ha='center', va="center", fontsize=14)
#plt.tight_layout()
plt.show()
```
